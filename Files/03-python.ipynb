## Basic Python Syntax

### Python Scalar Types

Python has the following built-in scalar types

| Type        | Example        | Description                                                  |
|-------------|----------------|--------------------------------------------------------------|
| ``int``     | ``x = 1``      | integers (i.e., whole numbers)                               |
| ``float``   | ``x = 1.0``    | floating-point numbers (i.e., real numbers)                  |
| ``complex`` | ``x = 1 + 2j`` | Complex numbers (i.e., numbers with real and imaginary part) |
| ``bool``    | ``x = True``   | Boolean: True/False values                                   |
| ``str``     | ``x = 'abc'``  | String: characters or text                                   |
| ``NoneType``| ``x = None``   | Special object indicating nulls                              |

### Arithmetic Operations
Python implements seven basic binary arithmetic operators:

| Operator     | Name           | Description                                            |
|--------------|----------------|--------------------------------------------------------|
| ``a + b``    | Addition       | Sum of ``a`` and ``b``                                 |
| ``a - b``    | Subtraction    | Difference of ``a`` and ``b``                          |
| ``a * b``    | Multiplication | Product of ``a`` and ``b``                             |
| ``a / b``    | True division  | Quotient of ``a`` and ``b``                            |
| ``a // b``   | Floor division | Quotient of ``a`` and ``b``, removing fractional parts |
| ``a % b``    | Modulus        | Integer remainder after division of ``a`` by ``b``     |
| ``a ** b``   | Exponentiation | ``a`` raised to the power of ``b``                     |

```python
a = 1
b = 2
(a + b, a - b, a / b, a // b, a % b, a ** b)
```

### Boolean Operations

When working with Boolean values, Python provides operators to combine the values using the standard concepts of "and", "or", and "not"

Predictably, these operators are expressed using the words `and`, `or`, and `not`:

```python
a = 5
b = 2
c = 3
(a == 5 or not b == 6) and not c == 3
```

### Identity and Membership Operators

Python also contains prose-like operators  to check for identity and membership.
They are the following:

| Operator      | Description                                       |
|---------------|---------------------------------------------------|
| ``a is b``    | True if ``a`` and ``b`` are identical objects     |
| ``a is not b``| True if ``a`` and ``b`` are not identical objects |
| ``a in b``    | True if ``a`` is a member of ``b``                |
| ``a not in b``| True if ``a`` is not a member of ``b``            |

Check for identity

```python
a = [1, 2, 3]
b = a
b is a, b is not a
```

Check for membership

```python
1 in [1, 2, 3], "foo" not in [1, 2, 3]
```

```python
i = 0
while True:
    print(i, end=" ")
    i += 1
    if i >= 10:
        break
```

### Functions

Python functions take position and keyword arguments

Positional arguments do not have a default value, keyword arguments do

You can optionally add type hints to functions to make them more readable and to help with debugging

```python
def fibonacci(n: int, a: int = 0, b: int = 1) -> list[int]:
    l = []
    while len(l) < n:
        a, b = b, a + b
        l.append(a)
    return l
```

```python
fibonacci(10)
```

```python
fibonacci(10, b=3, a=1)
```

Using the `*args` and `**kwargs` syntax you can pass an arbitrary number of arguments to a function

```python
def catch_all(*args, **kwargs):
    print("args =", args)
    print("kwargs = ", kwargs)
```

```python
catch_all(1, 2, 3, a=4, b=5)
```

```python
catch_all("a", keyword=2)
```

In a similar way, you can unpack arguments from a data structure

```python
inputs = (1, 2, 3)
keywords = {"pi": 3.14, "e": 1.67}

catch_all(*inputs, **keywords)
catch_all(1, 2, 3, pi=3.14, e=1.67)
```

### Classes

Let's define an abstract shape class and leave the implementation for child classes

```python
class Shape:

    def area(self) -> float:
        ...

    def circumference(self) -> float:
        ...
```

```python
shape = Shape()
```

Let's define a rectangle that inherits from `Shape`

```python
class Rectangle(Shape):
    def __init__(self, x: float, y: float) -> None:
        self.x = x
        self.y = x

    def area(self) -> float:
        return self.x * self.y

    def circumference(self) -> float:
        return 2 * self.x + 2 * self.y
```

We can initialize a rectangle and call its area and circumfrence methods

```python
rectangle = Rectangle(5.5, 3)
print(rectangle.area())
print(rectangle.circumference())
```

We can then further write a `Square` class that inherits from `Rectangle`

```python
class Square(Rectangle):
    def __init__(self, x: float) -> None:
        super().__init__(x, x)
```

We can initialize a square and call its area and circumfrence methods

```python
square = Square(4.5)
print(square.area())
print(square.circumference())
```

To check if a variable is an instance of class we can use the `isinstance` operator

```python
print(isinstance(square, Square))
print(isinstance(square, Shape))
```
