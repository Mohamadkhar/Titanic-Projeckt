### Subarrays as No-Copy Views

- NumPy array slices are returned as *views* rather than *copies* of the array data

````python
x2
````

````python
x2_sub = x2[:2, :2]
x2_sub
````

- If we modify the subarray, the original array changes!

````python
x2_sub[0, 0] = 99
````

````python
x2_sub
````

````python
x2
````

### Creating Copies of Arrays

- Despite the nice features of array views, it is sometimes useful to instead explicitly copy the data

````python
x2_sub_copy = x2[:2, :2].copy()
print(x2_sub_copy)
````

- If we now modify this subarray, the original array is not touched

````python
x2_sub_copy[0, 0] = 42
print(x2_sub_copy)
print(x2)
````

# Computation on NumPy Arrays: Universal Functions

- Computation on NumPy arrays can be very fast, or it can be very slow

- The key to making it fast is to use vectorized operations,  implemented through NumPy's *universal functions* (ufuncs)

- NumPy's ufuncs can be used to make repeated calculations on array elements much more efficient

## The Slowness of Loops

- Python is slow when many small operations are being repeated; for instance, looping over arrays to operate on each element

- The following function is a very typical implementation for computing a reciprocal in many programming languages

````python
def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output


values = np.random.randint(1, 10, size=5)
print(values)
compute_reciprocals(values)
````

- If we measure the execution time of this code for a large input, we see that this operation is very slow

````python
big_array = np.random.randint(1, 100, size=1_000_000)
%timeit compute_reciprocals(big_array)
````

*Question:* What do you think is the reason?

- The bottleneck here is the type checking and function dispatches that Python must do at each cycle of the loop

- Each time the reciprocal is computed, Python first examines the object's type and does a dynamic lookup of the correct function to use for that type

- In compiled code, this type specification would be known before the code executed and the result could be computed much more efficiently

## Introducing Ufuncs

- NumPy provides a convenient interface into this kind of statically typed, compiled routine

- This is known as a *vectorized* operation

- For simple operations, vectorization is as simple as using Python arithmetic operators directly on the array object

````python
print(compute_reciprocals(values))
print(1.0 / values)
````

- Looking at the execution time for our big array, we see that it completes orders of magnitude faster than the Python loop:

````python
%timeit compute_reciprocals(big_array)
%timeit (1.0 / big_array)
````

- Vectorized operations in NumPy are implemented via ufuncs and are extremely flexible

- For example, they can can also operate between two arrays:

````python
np.arange(5) / np.arange(1, 6)
````

- And ufunc operations are not limited to one-dimensional arrays.

````python
x = np.arange(9).reshape((3, 3))
2**x
````

**Rule of Thumb**

> Try to avoid manual loops when using NumPy arrays and instead use vectorized operations

## Exploring NumPy's Ufuncs

### Array Arithmetic

- Many of NumPy's ufuncs feel natural to use because they make use of Python's native arithmetic operators

- The standard operators can all be used:

````python
x = np.arange(4)
print("x      =", x)
print("x + 5  =", x + 5)
print("x - 5  =", x - 5)
print("x * 2  =", x * 2)
print("x / 2  =", x / 2)
print("x // 2 =", x // 2)
print("-x     =", -x)
print("x ** 2 =", x**2)
print("x % 2  =", x % 2)
````

- In addition, these can be strung together however you wish, and the standard order of operations is respected

````python
-((0.5 * x + 1) ** 2)
````

- All of these arithmetic operations are simply convenient wrappers around specific ufuncs built into NumPy

- For example, the `+` operator is a wrapper for the `add` ufunc

````python
np.add(x, 2)  # x + 2
````

- The following table lists how NumPy implements Python's standard arithmetic operators:

| Operator    | Equivalent ufunc  | Description                         |
|-------------|-------------------|-------------------------------------|
|`+`          |`np.add`           |Addition (e.g., `1 + 1 = 2`)         |
|`-`          |`np.subtract`      |Subtraction (e.g., `3 - 2 = 1`)      |
|`-`          |`np.negative`      |Unary negation (e.g., `-2`)          |
|`*`          |`np.multiply`      |Multiplication (e.g., `2 * 3 = 6`)   |
|`/`          |`np.divide`        |Division (e.g., `3 / 2 = 1.5`)       |
|`//`         |`np.floor_divide`  |Floor division (e.g., `3 // 2 = 1`)  |
|`**`         |`np.power`         |Exponentiation (e.g., `2 ** 3 = 8`)  |
|`%`          |`np.mod`           |Modulus/remainder (e.g., `9 % 4 = 1`)|

- More operators exist; we will now explore some of them for illustrative purposes

### Exponents and Logarithms

- Exponents and logathrims are implemented

````python
x = np.arange(1, 5)
print("x        =", x)
print("e^x      =", np.exp(x))
print("2^x      =", np.exp2(x))
print("3^x      =", np.power(3.0, x))
print("ln(x)    =", np.log(x))
print("log2(x)  =", np.log2(x))
print("log10(x) =", np.log10(x))
````

### Trigonometric Functions

- Also the common trigonometric functions are implemented

````python
theta = np.linspace(0, np.pi, 3)
print("theta      = ", theta)
print("sin(theta) = ", np.sin(theta))
print("cos(theta) = ", np.cos(theta))
print("tan(theta) = ", np.tan(theta))
````

# Aggregations: min, max, and Everything in Between

- The ufuncs we have seen up to now always return an array of the same shape as the array they received as input

- NumPy provides also several ufuncs to aggregate values

## Summing the Values in an Array

- For example, to sum the values of an array, use the `sum` function

````python
np.sum(np.random.random(100))
````

-  NumPy's version of sum is much faster than Python's built-in version

> While Python's built-in *operators* directly use NumPy functions, Python's built-in *functions* do not!

````python
%timeit sum(big_array)
%timeit np.sum(big_array)
````

## Minimum and Maximum

- Similarly, the `min` and `max` functions can be used to find the minimum value and maximum value of any given array

````python
np.min(big_array), np.max(big_array)
````

- The NumPy versions are again much faster

````python
%timeit min(big_array)
%timeit np.min(big_array)
````

- For `min`, `max`, `sum`, and several other ufuncs that aggregate values, a shorter syntax is to use methods of the array object itself

````python
big_array.min(), big_array.max(), big_array.sum()
````

### Multidimensional Aggregates

- One common type of aggregation operation is an aggregate along a row or column

- By default, NumPy aggregations will apply across all elements of a multidimensional array

````python
M = np.random.randint(0, 10, (3, 4))
M
````

````python
M.sum()
````

- Aggregation functions take an additional argument specifying the *axis* along which the aggregate is computed

````python
print("Minimum per column: ", M.min(axis=0))
print("Maximum per row:    ", M.max(axis=1))
````

- The `axis` keyword specifies the dimension of the array that will be *collapsed*, rather than the dimension that will be returned

````python
M
````

````python
M.shape
````

````python
M.sum(axis=0)
````

````python
M.sum(axis=0).shape
````
